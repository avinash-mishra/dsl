```python
# plot correlation matrix with numerical values
corr = dataset.corr()
sns.heatmap(corr)
```

```python
k = 3 #number of variables for heatmap
target_col = "Salary"
cols = dataset.corr().nlargest(k, target_col)[target_col].index
cm = dataset[cols].corr()
plt.figure(figsize=(8,4))
sns.heatmap(cm, annot=True, cmap = 'viridis')
```
